```python
import os
import subprocess

def create_unique_directory(root_path, base_folder_name, output_flag="json", model_name = 'BODY_25'):
    i = 0
    while os.path.exists(os.path.join(root_path, f"{base_folder_name}_{i}")):
        i += 1
    new_folder_name = f"{base_folder_name}_{i}"
    os.makedirs(os.path.join(root_path, new_folder_name))
    return new_folder_name

def process_videos_in_directory(directory_path, base_folder_name, output_flag="video", model_name = "BODY_25"):
    # List all video files in the specified directory
    video_files = [f for f in os.listdir(directory_path) if f.lower().endswith(('.mp4', '.avi', '.mov'))]

    # Set the root path (Google Drive root directory)
    root_path = "/content/drive/MyDrive"

    # Create a batch-specific folder
    batch_folder = create_unique_directory(root_path, base_folder_name)

    # Process each video file
    for video_file in video_files:
        # Get the video file name without extension
        file_name = os.path.splitext(video_file)[0]

        file_path = os.path.join(directory_path, video_file)
        # Create output directory using the video file name
        output_directory = os.path.join(root_path, batch_folder, file_name)

        # Execute OpenPose for each video
        command = f"cd /content/drive/MyDrive/openpose_copy && ./build/examples/openpose/openpose.bin --video \"{file_path}\" --display 0 --model_pose {model_name} --write_{output_flag} \"{output_directory}\""
        process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        stdout, stderr = process.communicate()

        print(f"Processed video: {video_file} | Output saved in directory: {output_directory}")
        print(f"Command output: {stdout.decode('utf-8')}")
        print(f"Command error: {stderr.decode('utf-8')}")
```
